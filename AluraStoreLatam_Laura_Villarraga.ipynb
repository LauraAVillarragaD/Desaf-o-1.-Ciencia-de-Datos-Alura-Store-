# 1. Análisis de facturación



```python
# Tarea calcular cual tienda vende más?

# Calculo de las ventas totales de cada una de las tiendas con la variable 'Precio' y el atributo sum
ventas_totales_tienda1 = tienda['Precio'].sum()
ventas_totales_tienda2 = tienda2['Precio'].sum()
ventas_totales_tienda3 = tienda3['Precio'].sum()
ventas_totales_tienda4 = tienda4['Precio'].sum()

# se usa el condicional if para determinar cual es la tienda que vende más
if ventas_totales_tienda1 > ventas_totales_tienda2 and ventas_totales_tienda1 > ventas_totales_tienda3 and ventas_totales_tienda1 > ventas_totales_tienda4:
  print(f'La tienda 1 es la tienda que más vende, con un total de {ventas_totales_tienda1}')
elif ventas_totales_tienda2 > ventas_totales_tienda1 and ventas_totales_tienda2 > ventas_totales_tienda3 and ventas_totales_tienda2 > ventas_totales_tienda4:
  print(f'La tienda 2 es la tienda que más vende, con un total de {ventas_totales_tienda2}')
elif ventas_totales_tienda3 > ventas_totales_tienda1 and ventas_totales_tienda3 > ventas_totales_tienda2 and ventas_totales_tienda3 > ventas_totales_tienda4:
  print(f'La tienda 3 es la tienda que más vende, con un total de {ventas_totales_tienda3}')
elif ventas_totales_tienda4 > ventas_totales_tienda1 and ventas_totales_tienda4 > ventas_totales_tienda2 and ventas_totales_tienda4 > ventas_totales_tienda3:
  print(f'La tienda 4 es la tienda que más vende, con un total de {ventas_totales_tienda4}')

# se muestra las ventas totales de cada una de las tiendas
print(f'Las ventas totales de la tienda 1 es igual a: {ventas_totales_tienda1}')
print(f'Las ventas totales de la tienda 2 es igual a: {ventas_totales_tienda2}')
print(f'Las ventas totales de la tienda 3 es igual a: {ventas_totales_tienda3}')
print(f'Las ventas totales de la tienda 4 es igual a: {ventas_totales_tienda4}')
```

##**Análisis**

La tienda 1 es la tienda que más vendio con un total de 1'150.880.400 pesos. Y la tienda 4 es la tienda que menos vende con un total de 1'038.375.700 pesos.
